```python
import psutil
import os
import json
from xml.dom import minidom
import xml.etree.ElementTree as ET
import zipfile
```

```python
for_case = input('1. Посмотреть информацию в консоле о логических дисках, именах, метке тома, размере и типе файловой системы\n'      
          '2. Работа с файлами\n'
          '3. Работа с JSON\n'
          '4. Работа с XML\n'
          '5. Работа с ZIP\n')


''''Задание 1
Вывести информацию в консоль о логических дисках, именах, метке тома, размере и типе файловой системы'''
if for_case == "1":
    #Цикл перебирает все разделы диска, возвращаемые функцией `disk_partitions()` из модуля `psutil`
    for partition in psutil.disk_partitions():
        print(f"Устройство: {partition.device}") #Выводит имя устройства (диска)
        print(f"Точка монтирования: {partition.mountpoint}")#Выводит точку монтирования раздела
        print(f"Тип файловой системы: {partition.fstype}")#Выводит тип файловой системы раздела
    # блок try-except для обработки исключения 
        try:
            partition_usage = psutil.disk_usage(partition.mountpoint)
        except PermissionError:
            # Пропускаем диски, к которым нет доступа
            continue
    print(f"Размер: {partition_usage.total / (1024 * 1024 * 1024):.2f} GB") # Выводит общий размер раздела в гигабайтах
    print(f"Занято: {partition_usage.used / (1024 * 1024 * 1024):.2f} GB") # Выводит используемое пространство
    print(f"Свободно: {partition_usage.free / (1024 * 1024 * 1024):.2f} GB")# Выводит свободное пространство

    
    
''''Задание 2:
Работа с файлами ( класс File, FileInfo, FileStream и другие)''''
if for_case == "2":
    # Создание файла
    with open("test.txt", "w") as file:
        pass
    print("Файл успешно создан!")
 
    # Запись в файл
    with open("test.txt", "w") as file:
        text = input("Введите текст: ")
        file.write(text)

    # Чтение файла
    with open("test.txt", "r") as file:
        text = file.read()
        print(text)

    # Удаление файла
    os.remove("test.txt")
    print("Файл успешно удален!")
    
    
    
''''Задание 3:
Работа с форматом JSON'''
if for_case == "3":
    #создается объект класса Cat с помощью вызова конструктора
    #и присваиваются значения для каждого атрибута с помощью функции input().
    #Далее вызывается метод dict() объекта Cat, который возвращает словарь 
    #с атрибутами и их значениями. Полученный словарь сохраняется в переменную data
    
    class Cat:
        def __init__(self):
            self.cat = cat
            self.breed = breed
            self.age = age
            self.gender = gender
    cat = input('Введите имя питомца: ')
    breed = input('Введите породу: ')
    age = input('Введите возраст: ')
    gender = input('Введите пол: ')

    data = Cat()
    
    #Эта часть кода открывает файл "datafile.json" для записи 
    #(если он не существует, то будет создан) и использует функцию json.dump() 
    #для записи словаря data.dict в формате JSON. Затем файл закрывается. 
    #Далее открывается тот же файл, но уже для чтения, и используется функция json.load() 
    #для чтения данных из файла и преобразования их в словарь. Полученный словарь сохраняется 
    #в переменную jsondata. После этого файл снова закрывается.'''
    
    with open("datafile.json", "w") as writefile:
        json.dump(data.__dict__, writefile)
    with open("datafile.json", "r") as readfile:
        jsondata = json.load(readfile)
    print("Файл успешно создан!")
    print(jsondata)
    os.remove('datafile.json')
    print("Файл успешно удален!")

    
    
'''Задание 4:
Работа с форматом XML'''  
if for_case == "4":  
    #создаем новый объект Document из модуля minidom, который используется для создания и обработки XML-документов
    root = minidom.Document()
    #создает элемент XML с именем «root» с помощью метода createElement()
    xml = root.createElement('root')
    #Затем он добавляет созданный элемент в качестве дочернего к корневому элементу
    root.appendChild(xml)

    productChild = root.createElement('Person') #родительский корневой элемент
    productChild.setAttribute('name', input("Введите имя: ")) #атрибуты в элементе
    productChild.setAttribute('Age', input("Введите возраст: "))
    productChild.setAttribute('city', input("Введите пол: "))

    xml.appendChild(productChild) # добавляем родительский элемент к XML

    xml_str = root.toprettyxml(indent="\t") #преобразуем XML в форматированное строковое представление 
    #с помощью метода toprettyxml и устанавливает уровень отступа «\t»

    save_path_file = "data_file.xml" #указывает путь и имя файла, в котором будет сохранен XML-документ
    #открываем + создаем файл
    with open(save_path_file, "w") as f:
        f.write(xml_str)
    #анализирует и сохраняет дерево в переменную
    tree = ET.parse('data_file.xml')
    #получает корневой элемент проанализированного XML-дерева,
    #используя метод getroot(), присваивая его переменной root
    root = tree.getroot()
    print(root) #корневой элемент
    print(root[0].attrib)#атрибуты первого дочернего элемента
    os.remove('data_file.xml')
    
    
''''Задание 5:
Создание zip архива, добавление туда файла, определение размера архива''''    
if for_case == "5":
    #создаем файл и заполняем его данными
    with open("example.txt", "w") as file:
        pass
    print("Файл успешно создан!")  
    with open("example.txt", "w") as file:
        text = input("Введите текст: ")
        file.write(text)
        
    # создание архива и добавление туда файла
    with zipfile.ZipFile('example.zip', 'w') as myzip:
        myzip.write('example.txt')
    print("Архив успешно создан!")
    
    
    # определение размера архива
    size = os.path.getsize('example.zip')
    print('Размер архива:',size)
    # вывод информации о файле
    with zipfile.ZipFile('example.zip', 'r') as myzip:
        for info in myzip.infolist():
            print(f"имя файла: {info.filename}")
            print( f"информация о дате создания(год, месяц, день, час, минуты, секунды): {info.date_time}" )
            print( f"информация о размере файла : {info.file_size} байт(а)")

    # разархивирование файла
    with zipfile.ZipFile('example.zip', 'r') as myzip:
        myzip.extractall()
    print("Файл разархивирован!")
    

    # удаление файла и архива
    os.remove('example.txt')
    os.remove('example.zip')
    print("Архив и файл успешно удалены!")
```
